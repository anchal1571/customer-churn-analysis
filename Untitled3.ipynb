```python
 # Plot churn distribution
churn_df['Churn'].value_counts().plot(kind='bar', color = ['lightgreen' , 'salmon'])
plt.title("Churn Distribution")
plt.xlabel("Churn")
plt.ylabel("Number of Customers")
plt.show()
  
```

```python
 churn_df.groupby(['gender', 'Churn']).size().unstack().plot(kind='bar', stacked=True)
plt.title("Churn by Gender")
plt.xlabel("Gender")
plt.ylabel("Number of Customers")
plt.show()
```

```python
churn_df.groupby(['Contract', 'Churn']).size().unstack().plot(kind='bar', stacked=True)
plt.title("Churn by Contract Type")
plt.xlabel("Contract Type")
plt.ylabel("Number of Customers")
plt.xticks(rotation=15)
plt.show()
```

```python
churn_df.boxplot(column='MonthlyCharges', by='Churn', grid=False)
plt.title("Monthly Charges by Churn")
plt.suptitle("")  # remove default subtitle
plt.xlabel("Churn")
plt.ylabel("Monthly Charges")
plt.show()
```

```python
churn_df[churn_df['Churn'] == 'Yes']['tenure'].plot(kind='hist', bins=30, color='salmon', edgecolor='black')
plt.title("Tenure Distribution of Churned Customers")
plt.xlabel("Tenure (months)")
plt.ylabel("Frequency")
plt.show()
```
